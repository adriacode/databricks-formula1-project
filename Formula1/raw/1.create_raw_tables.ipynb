```sql
USE CATALOG databricks_course_ws;
```

```sql
CREATE DATABASE IF NOT EXISTS f1_raw;
```

##### Create circuits table

```sql
CREATE OR REPLACE TABLE f1_raw.circuits
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/circuits.csv',
  format => 'csv',
  header => 'true',
  schema => 'circuitId INT, circuitRef STRING, name STRING, location STRING, country STRING, lat DOUBLE, lng DOUBLE, alt INT, url STRING'
)
```

##### Create races table

```sql
CREATE OR REPLACE TABLE f1_raw.races
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/races.csv',
  format => 'csv',
  header => 'true',
  schema => 'raceId INT, year INT, round INT, circuitId INT, name STRING, date DATE, time STRING, url STRING'
);
```

#### Create tables for JSON files

##### Create constructors table
- Single Line JSON
- Simple Structure

```sql
CREATE OR REPLACE TABLE f1_raw.constructors
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/constructors.json',
  format => 'json',
  schema => "constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRING"
);
```

##### Create drivers table
- Single Line JSON
- Complex structure

```sql
CREATE OR REPLACE TABLE f1_raw.drivers
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/drivers.json',
  format => 'json',
  schema => "driverId INT, driverRef STRING, number INT, code STRING, name STRUCT<forename: STRING, surname: STRING>, dob DATE, nationality STRING, url STRING"
);
```

##### Create results table
- Single Line Json
- Simple Structure

```sql
CREATE OR REPLACE TABLE f1_raw.results
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/results.json',
  format => 'json',
  schema => "resultId INT, raceId INT, driverId INT, constructorId INT, number INT, grid INT, position INT, positionText STRING, positionOrder INT, points FLOAT, laps INT, time STRING, milliseconds INT, fastestLap INT, rank INT, fastestLapTime STRING, fastestLapSpeed FLOAT, statusId INT"
);
```

##### Create pit stops tables
- Multi Line JSON
- Simple Structure


```sql
CREATE OR REPLACE TABLE f1_raw.pit_stops
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/pit_stops.json',
  format => 'json',
  schema => "driverId INT, duration STRING, lap INT, milliseconds INT, raceId INT, stop INT",
  multiLine => 'true'
);
```

#### Create tables for list of files

##### Create Lap Times Table
- CSV file
- Multiple files

```sql
CREATE OR REPLACE TABLE f1_raw.lap_times
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/lap_times',
  format => 'csv',
  schema => 'raceId INT, driverId INT, lap INT, position INT, time STRING, milliseconds INT')
```

##### Create Qualifying Table
- JSON file
- Multiline JSON
- Multiple files

```sql
CREATE OR REPLACE TABLE f1_raw.qualifying
AS SELECT * FROM read_files(
  'abfss://raw@formula1dl2026c.dfs.core.windows.net/qualifying',
  format => 'json',
  schema => "qualifyId INT, raceId INT, driverId INT, constructorId INT, number INT, position INT, q1 STRING, q2 STRING, q3 STRING",
  multiLine => 'true'
)
```
